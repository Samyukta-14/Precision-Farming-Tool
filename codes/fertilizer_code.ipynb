```python
import tkinter as tk
from tkinter import ttk
from tkinter import messagebox
import pandas as pd
from tkinter import font
from tkinter import scrolledtext
import sys
sys.path.append(r"C:\Users\pebbl\Downloads\Fertilizer_data.py")
from Fertilizer_data import fertilizer_dic

def fert_recommend():
    crop_name = crop_var.get()
    N = int(n_var.get())
    P = int(p_var.get())
    K = int(k_var.get())

    df = pd.read_csv(r"C:\Users\pebbl\Downloads\fertilizer (1).csv")

    nr = df[df['Crop'] == crop_name]['N'].iloc[0]
    pr = df[df['Crop'] == crop_name]['P'].iloc[0]
    kr = df[df['Crop'] == crop_name]['K'].iloc[0]

    n = nr - N
    p = pr - P
    k = kr - K
    temp = {abs(n): "N", abs(p): "P", abs(k): "K"}
    max_value = temp[max(temp.keys())]
    if max_value == "N":
        if n < 0:
            key = 'NHigh'
        else:
            key = "Nlow"
    elif max_value == "P":
        if p < 0:
            key = 'PHigh'
        else:
            key = "Plow"
    else:
        if k < 0:
            key = 'KHigh'
        else:
            key = "Klow"

    recommendation = fertilizer_dic[key]

    new_window = tk.Toplevel(root)
    new_window.title("Fertilizer Recommendation")

    window_width = 1500  
    window_height = 500  
    new_window.geometry(f"{window_width}x{window_height}")
    bold_font = font.Font(weight="bold")

    rec_text = scrolledtext.ScrolledText(new_window, wrap=tk.WORD)
    rec_text.insert(tk.INSERT, recommendation)
    rec_text.config(state=tk.DISABLED)  # Make it read-only
    rec_text.pack(fill=tk.BOTH, expand=True)

# Create the main window
root = tk.Tk()
root.title("Fertilizer Recommendation")

# Create labels and entry fields for N, P, K, and Crop Name
n_label = ttk.Label(root, text="N Value:")
n_label.grid(row=0, column=0, padx=10, pady=5)
n_var = tk.StringVar()
n_entry = ttk.Entry(root, textvariable=n_var)
n_entry.grid(row=0, column=1, padx=10, pady=5)

p_label = ttk.Label(root, text="P Value:")
p_label.grid(row=1, column=0, padx=10, pady=5)
p_var = tk.StringVar()
p_entry = ttk.Entry(root, textvariable=p_var)
p_entry.grid(row=1, column=1, padx=10, pady=5)

k_label = ttk.Label(root, text="K Value:")
k_label.grid(row=2, column=0, padx=10, pady=5)
k_var = tk.StringVar()
k_entry = ttk.Entry(root, textvariable=k_var)
k_entry.grid(row=2, column=1, padx=10, pady=5)

crop_label = ttk.Label(root, text="Crop Name:")
crop_label.grid(row=3, column=0, padx=10, pady=5)
crop_var = tk.StringVar()
crop_entry = ttk.Entry(root, textvariable=crop_var)
crop_entry.grid(row=3, column=1, padx=10, pady=5)

# Create a button to get fertilizer recommendation
recommend_button = ttk.Button(root, text="Get Fertilizer Recommendation", command=fert_recommend)
recommend_button.grid(row=4, column=0, columnspan=2, padx=10, pady=10)


root.mainloop()
```
